# SSIM

This code is a helper code to calculate the SSIM score between the heatmaps of the two different settings. It takes the heatmaps from the folders and simply calculates the SSIM between the two for all the given heatmaps. The results are printed and visualised in the end. 

```python
# PyTorch
import torch

from PIL import Image

# Third-Party Libraries
import numpy as np
import matplotlib.pyplot as plt
import os
from skimage.metrics import structural_similarity
```

```python
ssim_values = []
num_images = 2 # Needs to be changed to how many heatmaps there are
for i in range(num_images):
    # Take heatmaps from the folders
    img1 = np.array(torch.load(f'heatmaps/heatmap_clean_{i}.pt')).astype(np.float32)
    img2 = np.array(torch.load(f'heatmaps_dec/heatmap_clean_{i}.pt')).astype(np.float32)

    # Calculate score
    ssim_score = structural_similarity(img2, img1, multichannel=True, data_range=img2.max()-img2.min())
    print(ssim_score)
    ssim_values.append(ssim_score)

# Print summary and statistics
print("\nSummary Statistics:")
print(f"Average SSMI: {np.mean(ssim_values):.4f} ± {np.std(ssim_values):.4f}, Max:{np.max(ssim_values):.4f}, Min:{np.min(ssim_values):.4f}")
print(ssim_values)

plt.figure(figsize=(10, 5))
plt.bar(range(0, len(ssim_values)), ssim_values)
plt.axhline(y=np.mean(ssim_values), color='r', linestyle='--', label=f'Durchschnitt: {np.mean(ssim_values):.4f}')
plt.xlabel('Bild')
plt.ylabel('SSIM-Wert')
plt.ylim(0, 1)  
plt.show()
```
